```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import warnings #Elimina mensajes de alerta
warnings.simplefilter("ignore")
```

```python
# Cargar el conjunto de datos X_test y y_test
X_test = pd.read_pickle(r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\X_test.pkl')
y_test = pd.read_pickle(r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\y_test.pkl')
```

```python
# Cargar los modelos desde los archivos
modelo_regresion_logistica1 = joblib.load(r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\DATOS COMPLETOS\Algoritmo_regresion_logistica1.pkl')
modelo_arboles1 = joblib.load(r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\DATOS COMPLETOS\Algoritmo_arboles1.pkl')
modelo_bosques1 = joblib.load(r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\DATOS COMPLETOS\Algoritmo_bosques1.pkl')
modelo_gbm1 = joblib.load(r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\DATOS COMPLETOS\Algoritmo_gbm1.pkl')
```

```python
# Realizar las predicciones en el conjunto de prueba usando cada modelo
prediccion_rl1 = modelo_regresion_logistica1.predict(X_test)
prediccion_arboles1 = modelo_arboles1.predict(X_test)
prediccion_bosques1 = modelo_bosques1.predict(X_test)
prediccion_gbm1 = modelo_gbm1.predict(X_test)
```

```python
# Crear el DataFrame con los resultados comparativos
resultados_comparativos_1 = pd.DataFrame({
    'Real': y_test,
    'Predicción Regresión Logística': prediccion_rl1,
    'Predicción Árboles de Decisión': prediccion_arboles1,
    'Predicción Bosques Aleatorios': prediccion_bosques1,
    'Predicción GBM': prediccion_gbm1
})

# (Opcional) Guardar en un archivo Excel
resultados_comparativos_1.to_excel('resultados_filtrados_1.xlsx', index=False)

# Mostrar todo el DataFrame sin filtrado
resultados_comparativos_1
```

```python
# Contar la cantidad de 1 y 0 en cada columna
conteo_unos_1 = resultados_comparativos_1.apply(lambda x: (x == 1).sum(), axis=0)
conteo_ceros_1 = resultados_comparativos_1.apply(lambda x: (x == 0).sum(), axis=0)

# Mostrar la cantidad de unos y ceros por columna
print("\nCantidad de unos por columna:")
print(conteo_unos_1)

print("\nCantidad de ceros por columna:")
print(conteo_ceros_1)
```

```python
# Listado de los modelos que deseas cargar
modelos1 = [
    r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\DATOS COMPLETOS\Algoritmo_regresion_logistica1.pkl',
    r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\DATOS COMPLETOS\Algoritmo_arboles1.pkl',
    r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\DATOS COMPLETOS\Algoritmo_bosques1.pkl',
    r'C:\Users\Dani\Documents\DIPLOMADO\PROYECTO\DATOS COMPLETOS\Algoritmo_gbm1.pkl'
]

# Colores y etiquetas para cada modelo
colores = ['blue', 'green', 'red', 'purple']
etiquetas = ['Regresión Logística', 'Árboles de Decisión', 'Bosques Aleatorios', 'GBM']

# Crear la figura con 4 subgráficas, cada una con dos partes
fig, axes = plt.subplots(4, 2, figsize=(12, 18))  # 4 modelos y 2 columnas (real y predicción)

# Iterar sobre los modelos para cargar y graficar las predicciones
for i, (modelo_path, color, etiqueta) in enumerate(zip(modelos1, colores, etiquetas)):
    # Cargar el modelo
    modelo_cargado1 = joblib.load(modelo_path)
    
    # Realizar las predicciones con el modelo cargado
    y_pred = modelo_cargado1.predict(X_test)
    
    # Subgráfico 1: Valores Reales
    ax1 = axes[i, 0]  # Primera columna para los valores reales
    ax1.scatter(range(len(y_test)), y_test, color='black', label='Valor Real', alpha=0.8)
    ax1.set_title(f'Valores Reales - {etiqueta}')
    ax1.set_xlabel('Índice de Ejemplo')
    ax1.set_ylabel('Valor')
    ax1.legend()
    
    # Subgráfico 2: Predicciones
    ax2 = axes[i, 1]  # Segunda columna para las predicciones
    ax2.scatter(range(len(y_pred)), y_pred, color=color, label=f'{etiqueta} (Predicción)', alpha=0.6)
    ax2.set_title(f'Predicciones - {etiqueta}')
    ax2.set_xlabel('Índice de Ejemplo')
    ax2.set_ylabel('Valor')
    ax2.legend()

# Agregar un título general para toda la figura
fig.suptitle('Comparación de Valores Reales y Predicciones Modelos sin Balanceo', fontsize=16, y=1.02)

# Ajustar el espacio entre las subgráficas
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```
